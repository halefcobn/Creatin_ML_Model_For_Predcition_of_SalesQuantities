```python
import pandas as pd
import sys

train_data = pd.DataFrame()
test_data = pd.DataFrame()
tested_products = set()
count = 0

results = []



for (product_id, product_code), group_df in grouped_data:
    if group_df['period_id'].nunique() < 10:
        # Total enumeration of period ID is less than 10, move to the next product
        continue

    if (product_id, product_code) in tested_products:
        # Skip testing if the product has already been tested
        continue

    for period_id in range(2, group_df['period_id'].max() + 1):
        test_df = group_df[group_df['period_id'] == period_id]
        train_df = group_df[group_df['period_id'] != period_id]

        train_data = train_data.append(train_df)
        test_data = test_data.append(test_df)

        X_train = train_data.drop("net_quantity", axis=1)
        X_test = test_data.drop("net_quantity", axis=1)
        y_train = train_data.net_quantity
        y_test = test_data.net_quantity

        # Define the parameter grid for GridSearchCV
        param_grid = {
            'objective': ['reg:squarederror'],
            'learning_rate': [0.1, 0.01],
            'max_depth': [3, 5],
            'min_child_weight': [1, 3],
            'subsample': [0.8, 1.0],
            'colsample_bytree': [0.8, 1.0],
            'eval_metric': ['rmse']
        }

        # Create an XGBoost regressor
        xgb_model = xgb.XGBRegressor()

        # Create a GridSearchCV object
        grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='neg_root_mean_squared_error', cv=5)

        # Fit the GridSearchCV object to the training data
        grid_search.fit(X_train, y_train)

        # Get the best parameters and best score
        best_params = grid_search.best_params_
        best_score = grid_search.best_score_

        # Create an XGBoost model with the best parameters
        best_model = xgb.XGBRegressor(**best_params)

        # Train the XGBoost model with the best parameters
        best_model.fit(X_train, y_train)

        # Predict the target variable for the test data
        y_pred_train = best_model.predict(X_train)
        y_pred_test = best_model.predict(X_test)

        # Calculate evaluation metrics
        mse_train = mean_squared_error(y_train, y_pred_train)
        rmse_train = np.sqrt(mse_train)
        r2_train = r2_score(y_train, y_pred_train)

        mse_test = mean_squared_error(y_test, y_pred_test)
        rmse_test = np.sqrt(mse_test)
        r2_test = r2_score(y_test, y_pred_test)

        if r2_train >= 0.55 and r2_test >= 0.55:
            count += 1
            print(count)
            # Both training and testing R-squared values are greater than 0.55
            print("Product ID:", product_id)
            print("Product Code:", product_code)
            print("Period ID:", period_id)
            print("Best Parameters:", best_params)
            print("Best Score:", best_score)
            print("Training Set:")
            print("RMSE:", rmse_train)
            print("R2 Score:", r2_train)
            print("Test Set:")
            print("RMSE:", rmse_test)
            print("R2 Score:", r2_test)
            print("High R-squared value reached. Moving to the next period ID.")

            # Generate a prediction for period 2
            X_period_2 = test_data.drop("net_quantity", axis=1)
            y_pred_period_2 = best_model.predict(X_period_2)
            last_true_value = y_test.iloc[-1]
            last_predicted_value = y_pred_period_2[-1]
            print("Period 2 Prediction: True Value =", last_true_value, "Predicted Value =", round(last_predicted_value))

            results.append({
                'Product ID': product_id,
                'Product Code': product_code,
                'Period ID': period_id,
                'Best Parameters': best_params,
                'Best Score': best_score,
                'Training Set RMSE': rmse_train,
                'Training Set R2 Score': r2_train,
                'Test Set RMSE': rmse_test,
                'Test Set R2 Score': r2_test,
                'Period 2 True Value': last_true_value,
                'Period 2 Predicted Value': round(last_predicted_value)
            })

            break

    # Add the tested product ID and product code to the set of tested products
    tested_products.add((product_id, product_code))

# Create a DataFrame from the results list
results_df = pd.DataFrame(results)

# Save the DataFrame to an Excel file
results_df.to_excel('outputprediction.xlsx', index=False)
```
